```python
import torch
from torch import nn
from torch.utils.data import DataLoader
from torchvision import datasets
from torchvision.transforms import ToTensor
```

```python
# Download training data from open datasets.
training_data = datasets.FashionMNIST(
    root="data",
    train=True,
    download=True,
    transform=ToTensor(),
)

# Download test data from open datasets.
test_data = datasets.FashionMNIST(
    root="data",
    train=False,
    download=True,
    transform=ToTensor(),
)
```

```python
batch_size = 64

# Create data loaders.
train_dataloader = DataLoader(training_data, batch_size=batch_size)
test_dataloader = DataLoader(test_data, batch_size=batch_size)

for X, y in test_dataloader:
    print(f"Shape of X [N, C, H, W]: {X.shape}")
    print(f"Shape of y: {y.shape} {y.dtype}")
    break
```

```python
# https://discuss.pytorch.org/t/should-i-use-model-eval-when-i-freeze-batchnorm-layers-to-finetune/39495/4

normalization_layers = [
    nn.BatchNorm1d, nn.BatchNorm2d, nn.BatchNorm3d, nn.LazyBatchNorm1d, nn.LazyBatchNorm2d, nn.LazyBatchNorm3d,
    nn.LazyBatchNorm3d, nn.GroupNorm, nn.SyncBatchNorm, nn.InstanceNorm1d, nn.InstanceNorm2d, nn.InstanceNorm3d,
    nn.LazyInstanceNorm1d, nn.LazyInstanceNorm2d, nn.LazyInstanceNorm3d, nn.LayerNorm, nn.LocalResponseNorm
]

def is_norm_layer(layer):
    """
    Check if a layer is a normalization layer.

    Args:
        layer (torch.nn.Module): A layer from the model.

    Returns:
        bool: True if the layer is a normalization layer, False otherwise.
    """
    for norm_layer in normalization_layers:
        if isinstance(layer, norm_layer):
            return True
    return False

def set_norm_eval_and_freeze(module):
    """
    Recursively sets normalization layers to evaluation mode and freezes their parameters.

    Args:
        module (torch.nn.Module): A module from the model.
    """
    for layer in module.children():
        if is_norm_layer(layer):
            # Set the layer to evaluation mode
            layer.eval()
            # Freeze the layer parameters
            for param in layer.parameters():
                param.requires_grad = False
                
        # Recursively apply to child modules
        set_norm_eval_and_freeze(layer)
```

```python
# The trainable parameters of the `model`, by reference
# Lies in `device` (same as each `param`)
trainable_parameters = [param for param in model.parameters() if param.requires_grad]
```

```python
# Lies in `device` (same as each `param`)
round_start_trainable_parameters = [param.detach().clone() for param in trainable_parameters]
```

```python
# Lies in `device` (same as each `param`)
aggregated_trainable_parameters = [torch.zeros_like(param) for param in trainable_parameters]
```

```python
@torch.no_grad
def zero_parameters(parameters):
    """
    Sets all parameters in the given iterable to zero.

    This function operates in-place and modifies the parameters directly.
    The @torch.no_grad() decorator ensures that this operation is not tracked 
    by autograd, preventing unnecessary computation and memory usage.

    Args:
        parameters (iterable): An iterable of parameters to be zeroed.
    """
    for param in parameters:
        param.zero_()
```

```python
zero_parameters(aggregated_trainable_parameters)
```

```python
@torch.no_grad
def copy_parameters(from_parameters, to_parameters):
    """
    Copies the values from one set of parameters to another.

    This function operates in-place and modifies the `to_parameters` directly.
    The @torch.no_grad() decorator ensures that this operation is not tracked 
    by autograd, preventing unnecessary computation and memory usage.

    Args:
        from_parameters (iterable): An iterable of source parameters to copy from.
        to_parameters (iterable): An iterable of destination parameters to copy to.
    """
    for from_param, to_param in zip(from_parameters, to_parameters):
        to_param.copy_(from_param)
```

```python
copy_parameters(
    from_parameters=trainable_parameters, 
    to_parameters=round_start_trainable_parameters
)
```

```python
# -------------- TRAIN ONE CLIENT
```

```python
# incorporate the client's learnt weights
with torch.no_grad():
    for param, aggr_param in zip(trainable_parameters, aggregated_trainable_parameters):
        # TODO: n - current client's number of samples
        n = 1
        aggr_param.add_(n * param)
```

```python
# divide with the number of total samples seen in this round (all clients)
# TODO: N?
N = 1
for param in aggregated_trainable_parameters:
    param.div_(N)
```

```python
# -------------- set 
```

```python
copy_parameters(
    from_parameters=aggregated_trainable_parameters, 
    to_parameters=trainable_parameters
)
```

```python
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=1e-3)
```

```python
optimizer
```

```python
for i, (X, y) in enumerate(train_dataloader):
    X, y = X.to(device), y.to(device)
    
    pred = model(X)
    loss = loss_fn(pred, y)
    
    loss.backward()
    optimizer.step()
    optimizer.zero_grad()
    
    if i % 100 == 0:
        print(f"step: {i} loss: {loss.item()}")
```
